```python
data_2=data2[(data['z'] <= 0.055)]
```

```python
ax_2 = data_2.plot.scatter('x','z')
```

```python
data_2=data_2.drop(columns=['y','type','NaN'])
```

```python
data_2
```

```python
data_2=data_2.sort_values('x')
data_2
```

```python
x=0
idx_array=np.zeros(252)
i=0
while x<0.126:
    x1=x
    x2=x+0.001
    df=data_2[(data_2['x']>x1) & (data_2['x']<=x2)]
    idx=df['z'].idxmax()
    idx_array[i]=idx
    i=i+1
    x=x+0.0005
```

```python
top_par=data_2.loc[idx_array]
ax_3 = top_par.plot.scatter('x','z')
```

```python
par_fit=top_par[(top_par['x']>0.01) & (top_par['x']<0.04)]
x_values=par_fit[['x']]
z_values=par_fit[['z']]
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(x_values,z_values)
model.coef_
np.rad2deg(np.arctan(model.coef_))
```

```python
ax = data2.plot.scatter('x','z')
ax.plot(x_values, model.predict(x_values), color='red')
```
